```python
import pandas as pd

from sklearn.preprocessing import PowerTransformer
```

```python
df = pd.read_csv('../data/intermediate_data/df.csv')
```

```python
log = PowerTransformer()
log.fit(df[['population']])

df['log_population'] = log.transform(df[['population']])
```

```python
log = PowerTransformer()
log.fit(df[['amount_tsh']])

df['log_amount_tsh'] = log.transform(df[['amount_tsh']])
```
